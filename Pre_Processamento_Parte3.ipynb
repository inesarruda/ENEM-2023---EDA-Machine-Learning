## Pré Processamento

#### Nessa etapa será feita a divisão entre os dados de treino e teste

#### Como temos uma grande diversidade dentro da amostra, é mais proveitoso pegar uma **amostra estratificada**

#### Para que a amostra seja o mais fidedígino possível, a amostra será na veradade uma amostra estratificada levando em consideração as colunas:
 - "Sexo"
 - "Renda Familiar"  
 - "Estado de Origem"

##### Será feita uma divisão estratificada de dados de treino e teste tanto para o dataframe categórico, quanto para o dataframe numérico. 
##### Um dos motivos disso desse split desde antes da análise exploratória é para evitar alguns problemas como data leakage, que é o vazamento dos dados de treino para os dados de teste ao realizar quaisquer tipo de pré processamento. Além disso, durante a análise de dados, estaremos o mais próximo possível da realidade de um projeto cotidiano, onde não temos acesso aos dados de teste, pois eles geralmente "não aconteceram" ou não temos acesso.

Além disso, será aplicada uma breve etapa de feature engineering e em seguida os dataframes serão salvos em parquet para utilizar na etapa de análise de dados (EDA) e da criação dos modelos de Machine Learning



```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

#### Aplicação do train_test_split no dataframe numérico 

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Criando a variável estratificada combinada
#df["estrato"] = df["TP_SEXO"] + "_" + df["renda_familiar"]
df_numerico['estrato'] = df_numerico["TP_SEXO"].astype('str') + "_" + df_numerico["renda_familiar"].astype('str') + "_" + df_numerico['SG_UF_ESC'].astype('str')

# Divisão estratificada com train_test_split
df_train_num, df_test_num = train_test_split(df_numerico, test_size=0.2, random_state=42, stratify=df_numerico["estrato"])

# Verificando se a distribuição foi mantida
print("Distribuição no dataset original:")
print(df_numerico["estrato"].value_counts(normalize=True))

print("\nDistribuição no treino:")
print(df_train_num["estrato"].value_counts(normalize=True))

print("\nDistribuição no teste:")
print(df_test_num["estrato"].value_counts(normalize=True))
```

#### Aplicação do train_test_split no dataframe categórico 

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Criando a variável estratificada combinada
#df["estrato"] = df["TP_SEXO"] + "_" + df["renda_familiar"]
df_cat['estrato'] = df_cat["TP_SEXO"].astype('str') + "_" + df_cat["renda_familiar"].astype('str') + "_" + df_cat['SG_UF_ESC'].astype('str')

# Divisão estratificada com train_test_split
df_train_cat, df_test_cat = train_test_split(df_cat, test_size=0.2, random_state=42, stratify=df_cat["estrato"])

# Verificando se a distribuição foi mantida
print("Distribuição no dataset original:")
print(df_cat["estrato"].value_counts(normalize=True))

print("\nDistribuição no treino:")
print(df_train_cat["estrato"].value_counts(normalize=True))

print("\nDistribuição no teste:")
print(df_test_cat["estrato"].value_counts(normalize=True))
```

#### Vamos passar os dataframes de treino e teste, tanto numérico quanto categóricos para parquet. Durante parte de tratamento de nulos e análise de dados iremos usar apenas os dados de treino para evitar qualquer possibilidade de data leakege

### Feature Engineering

#### Também vamos fazer um pouco de feature engineering parar criar variáveis uma variável de Média Geral

No futuro de atualizações deste projeto, algumas outras colunas serão criadas, como renda per capita, mas ficará para o próximo commit

```python
def feature_eginerring(df):
    df['MEDIA_GERAL'] = (df['NU_NOTA_CN'] + df['NU_NOTA_CH'] + df['NU_NOTA_LC'] + df['NU_NOTA_MT'] + df['NU_NOTA_REDACAO'])/5   
    return df
```

Criando as colunas

```python
df_train_cat = feature_eginerring(df_train_cat)
df_test_cat = feature_eginerring(df_test_cat)
df_train_num = feature_eginerring(df_train_num)
df_test_num = feature_eginerring(df_test_num)
```

Salvando os dataframes em parquet

```python
df_train_cat.to_parquet("df_train_cat.parquet", compression = 'gzip')
df_test_cat.to_parquet("df_test_cat.parquet", compression = 'gzip')

df_train_num.to_parquet("df_train_num.parquet", compression = 'gzip')
df_test_num.to_parquet("df_test_num.parquet", compression = 'gzip')
```
